```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pathlib
%matplotlib inline
```

```python
data_dir = pathlib.Path("../data/helen/helen_1").resolve()
```

```python
fig, axes = plt.subplots(5, 5)
for path, axis in zip(data_dir.glob("*.jpg"), axes.flatten()):
    img = cv2.imread(str(path))
    axis.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axis.axis('off')
fig.savefig("face_grid.pdf")
```
